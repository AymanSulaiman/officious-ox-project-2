```python
import ml_outputs as mo
import pandas as pd
```

```python
df = pd.read_csv('cleaned_tvg_cd_data.csv')
df.tail()
```

```python
features = [
    'tvghorseweight',            # 0
    'tvghorseage',               # 1
    'tvghorsedaysoff',           # 2
    'tvghorsenumberofwins',      # 3
    'tvghorsepowerrating',       # 4
    'tvghorseaverageclassrating' # 5
]

X= df[features]
y= df.tvghorseaveragespeed.values.reshape(-1,1)

X_train, X_test, y_train, y_test = mo.splitter(X=X, y=y).splitting_function()
```

```python
X_train_scaled, X_test_scaled = mo.scaler.X_scaler_function(X_train,X_test)
```

```python
y_train_scaled, y_test_scaled = mo.scaler.y_scaler_function(y_train,y_test)
```

```python
lin_reg = mo.regression_models.linear_regression(X_train_scaled, 
                                                 X_test_scaled,
                                                 y_train_scaled, 
                                                 y_test_scaled)
```
