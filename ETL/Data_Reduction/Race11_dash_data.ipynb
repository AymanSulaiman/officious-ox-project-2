```python
import pandas as pd
```

```python
fullcard = pd.read_csv('../data/Op_200502_h.csv')
```

```python
fullcard
```

```python
race11 = fullcard[fullcard.RCRace == 11]
```

```python
pd.set_option('display.max_columns', 100)
race11
```

```python
for col in race11.columns: 
    print(col) 
```

```python
race11_condensed = race11[[
'PostPosition','Horse','Trainer','Jockey',
'Sire',
'Dam',
'Owner',
'SalePrice',
'LastRaced',
'Earnings',
'Starts',
'Wins',
'Places',
'Shows',
'WEarnings',
'WStarts',
'WWins',
'WPlaces',
'WShows',
]].copy()
```

```python
race11_condensed
```

```python
race11_condensed['SalePrice'] = race11_condensed['SalePrice']*10
```

```python
race11_condensed['SalePrice'] = race11_condensed['SalePrice'].apply(lambda x: "${:.1f}k".format((x/1000)))
```

```python
race11_data = race11_condensed
```

```python
race11_data
```

```python
race11_data.to_json(r'race11.json')
```
