```python
path = r'C:Horse_Data' # use your path
all_files = glob.glob(path + "/*.csv")

my_list = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    my_list.append(df)

race_data = pd.concat(my_list, axis=0, ignore_index=True)
```

```python
race_data
```

```python
race_data_new = race_data.rename(columns = {"R_RCTrack":"tvgtrackcode", 
                                          "R_RCDate": "racedate", 
                                          "R_RCRace": "race", 
                                          "B_ProgNum": "bettinginterestnumber"})
```

```python
race_data_new
```

```python
race_data_new.to_csv("race_data.csv")
```
